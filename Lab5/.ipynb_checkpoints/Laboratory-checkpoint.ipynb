```python
'''Natural Language Interface'''

''' Classes'''

class Cat:
    def __init__(self, name, color, language):
        self.name = name
        self.color = color
        self.language = language
    felix = Cat("felix", "black", "meow!")
    
class Dog:
    def __init__(self, name, color, language):
        self.name = name
        self.color = color
        self.language = language
    
    def bark(self):
        print(self.language)
fido = Dog("Fido", "brown", "woof!")
fido.bark()
```

```python
'''Inheritance'''
class Animal:
    legs = 4
    def __init__(self, name, color, language):
        self.name = name
        self.color = color
        self.language = language
    
class Cat(Animal):
    def purr(self):
        print(self.language)
        
class Dog(Animal):
    def bark(self):
        print(self.language)
    
felix = Cat("felix", "black", "meow!")
felix.purr()

fido = Dog("Fido", "brown", "woof!")
fido.bark()
    
```

```python
''' Properties '''
class Pizza:
    def __init__(self, topping):
        self.topping = topping
        
    @property
    def pineapple_allowed(self):
        return False
pizza = Pizza(["cheese", "tomato"])
print(pizza.pineapple_allowed)

pizza.pineapple_allowed = True
print(pizza.pineapple_allowed)
```

```python
''' Properties can also be set by defining setter/getter '''
class Pizza:
    def __init__(self, toppings):
        self.toppings = toppings
        self._pineapple_allowed = False
    
    ''' Define property'''
    @property
    def pineapple_allowed(self):
        return self._pineapple_allowed
    
    ''' Define property setter'''
    @pineapple_allowed.setter
    def pineapple_allowed(self, value):
        if value:
            password = input("Enter the password")
            if password == "Abracadabra":
                self._pineapple_allowed = value
            else:
                raise ValueError("Alert! Intruder!")
                
pizza = Pizza(["cheese", "tomato"])
print(pizza.pineapple_allowed)
```
